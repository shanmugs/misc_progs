## Pre-Processing

```r
# Load tm package
library(tm)
```

```r
# Create Corpus
corpus = VCorpus(VectorSource(emails$email))
```

```r
corpus[[1]]$content
```

```r
# Convert to lower-case
corpus = tm_map(corpus, tolower)
```

```r
# Necessary pre-process data
corpus = tm_map(corpus, PlainTextDocument)
```

```r
# Remove punctuation
corpus = tm_map(corpus, removePunctuation)

# Remove stop words
corpus = tm_map(corpus, removeWords, stopwords("english"))

# Stem document
corpus = tm_map(corpus, stemDocument)
```

```r
# Examine first email
corpus[[1]]$content
```

### Bag of Words

```r
# Create matrix
dtm = DocumentTermMatrix(corpus)
dtm
```

So we want to remove the terms that don't appear too often in our data set, and we'll do that using the removeSparseTerms function. And we're going to have to determine the sparsity, so we'll say that we'll remove any term that doesn't appear
in at least 3% of the documents. To do that, we'll pass 0.97 to removeSparseTerms.

```r
# Remove sparse terms
dtm = removeSparseTerms(dtm, 0.97)
dtm
```

We can see that we've decreased the number of terms (22164 to 788), which is a much more reasonable number.

```r
# Create data frame
labeledTerms = as.data.frame(as.matrix(dtm))

head(labeledTerms)
```

```r
# Add in the outcome variable
labeledTerms$responsive = emails$responsive

head(labeledTerms)
```

```r
str(labeledTerms)
```

So as we expect, there are an awful lot of variables, 789 in total. 788 of those variables are the frequencies of various words in the emails, and the last one is responsive, the outcome variable.

### Split the data

```r
library(caTools)
```

```r
set.seed(144)

spl = sample.split(labeledTerms$responsive, 0.7)

train = subset(labeledTerms, spl == TRUE)
test = subset(labeledTerms, spl == FALSE)
```

### Build a CART model

```r
library(rpart)
library(rpart.plot)
```

```r
# Classification and Regression Trees (CART)
emailCART = rpart(responsive~., data=train, method="class")

# Plot Tree
prp(emailCART)
```

So we can see at the very top is the word California. If California appears at least twice in an email, we're going to take the right part over here and predict that a document is responsive. It's somewhat unsurprising that California shows up, because we know that Enron had a heavy involvement in the California energy markets.

So further down the tree, we see a number of other terms that we could plausibly expect to be related to energy bids and energy scheduling, like system, demand, bid, and gas. Down here at the bottom is Jeff, which is perhaps a reference to Enron's CEO, Jeff Skillings, who ended up actually being jailed for his involvement in the fraud at the company.

```r
# Make Predictions
pred = predict(emailCART, newdata=test)

# Look at the first 10 rows
pred[1:10,]
```

So the **left column** here is the predicted probability of the document being **non-responsive**. And the **right column** is the predicted probability of the document being **responsive**. They sum to 1.

So in our case, **we want to extract the predicted probability of the document being responsive**. We're looking for the rightmost column. We'll create an object called *pred.prob*. And we'll select the rightmost or second column.

```r
# pred.prob will contain our test set predicted probabilities.
pred.prob = pred[,2]
```

```r
# Confusion Matrix
cmCA = table(test$responsive, pred.prob >= 0.5)
cmCA
```

We correctly identified 25 a responsive document. In 20 cases, we identified a document as responsive, but it was actually non-responsive.

The rows are labeled with the actual outcome, and the columns are labeled with the predicted outcome.

                      Predict 0       Predict 1
        Actual 0    True Negative   False Positive
        Actual 1    False Negative  True Positive

        cmCA = [1][3]
               [2][4]

```r
# Compute Classification and Regression Trees Accuracy
accurCA = sum(diag(cmCA))/sum(cmCA)
paste("Accuracy CART:", round(accurCA,digits=4))
```

```r
# Compute Baseline Accuracy
Bl = table(test$responsive)

accurBl = max(Bl)/sum(Bl)
paste("Accuracy Baseline:", round(accurBl,digits=4))
```

### Relative Operating Characteristic curve (ROC)

```r
library(ROCR)
```

```r
# Plot ROC

predROCR = prediction(pred.prob, test$responsive)

perfROCR = performance(predROCR, "tpr", "fpr")

plot(perfROCR, colorize=TRUE, main="ROC curve")
```

```r
# Area Under the ROC Curve (AUC)
AUC = as.numeric(performance(predROCR, "auc")@y.values)

paste("AUC CART:", round(AUC,digits=4))
```

We can see that we have an AUC in the test set of 79.4%, which means that our model can differentiate between a randomly selected responsive and non-responsive document about 80% of the time.
