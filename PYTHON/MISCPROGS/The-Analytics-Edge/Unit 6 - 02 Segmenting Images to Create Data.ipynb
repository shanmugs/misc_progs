# Seeing the Big Picture: Segmenting Images to Create Data

<img src="images/matrix.png"/>

```r
# Load the dataset
flower = read.csv("data/flower.csv", header=FALSE) # Our data doesn't have a header

head(flower,3)
```

```r
str(flower)
```

Actually, R treats the rows as observations and the columns as variables. 

```r
# Change the data type to matrix
flowerMatrix = as.matrix(flower)

head(flowerMatrix,3)
```

```r
str(flowerMatrix)
```

Now if we look at the structure of the flower matrix, we realize that we have 50 rows and 50 columns. What this suggests is that the resolution of the image is 50 pixels in width and 50 pixels in height.

```r
# Turn matrix into a vector
flowerVector = as.vector(flowerMatrix)

str(flowerVector)
```

**Converting the data to a matrix and then to the vector is a crucial step.**

```r
# Compute distances
distance = dist(flowerVector, method = "euclidean")
```

```r
# Hierarchical clustering
clusterIntensity = hclust(distance, method="ward.D")
```

```r
# Plot the dendrogram
plot(clusterIntensity)

# Create 3 clusters on the dendrogram
rect.hclust(clusterIntensity, k = 3, border = "red")  # make rectangles
```

```r
# Divide the dataset into 3 clusters
flowerClusters = cutree(clusterIntensity, k = 3)
```

```r
# Find mean intensity values
tapply(flowerVector, flowerClusters, mean)
```

What we obtain is that the first cluster has a mean intensity value of 0.08, which is closest to zero, and this means that it corresponds to the darkest shape in our image. And then the third cluster here, which is closest to 1, corresponds to the fairest shade.

```r
library(xtable)
```

```r
flowerClusters
```

```r
# Compare two groups using a statistical measure
z = tapply(flowerVector, flowerClusters, mean)
z
```

### Plot the image and the clusters

```r
# Set the dimensions of the image
dim(flowerClusters) = c(50,50)
image(flowerClusters, axes = FALSE)
```

### Original image

```r
# Original Image
image(flowerMatrix,axes=FALSE,col=grey(seq(0,1,length=256)))
```

# Let's try this with an MRI image of the brain

```r
healthy = read.csv("data/healthy.csv", header=FALSE)
```

```r
healthyMatrix = as.matrix(healthy)

str(healthyMatrix)
```

```r
# Plot image
image(healthyMatrix,axes=FALSE,col=grey(seq(0,1,length=256)))
```

# Hierarchial Clustering

```r
# Hierarchial clustering

healthyVector = as.vector(healthyMatrix)

distance = dist(healthyVector, method = "euclidean")
```

```r
# We have an error - why?
str(healthyVector)
```

The healthy vector has 365,636 elements. Let's call this number n. And remember, from our previous video, that for R to calculate the pairwise distances, it would actually need to calculate n*(n-1)/2 and then store them in the distance matrix.

```r
n = 365636

storage = n*(n-1)/2

storage
```

For n =365636, it's 67 billion values that we're asking R to store in a matrix. The bad news now is that we cannot use hierarchical clustering. It is impossible for us to use hierarchical clustering because of the high resolution of our image.

The k-means clustering algorithm aims at partitioning the data into k clusters, in a way that each data point belongs to the cluster whose mean is the nearest to it.

# K-means Clustering

<img src="images/kmeans.png"/>

```r
# (1) Specify number of clusters
k = 5
```

```r
# (2), (3)... Run k-means

set.seed(1)

KMC = kmeans(healthyVector, centers = k, iter.max = 1000)

str(KMC)
```

```r
# Extract clusters

healthyClusters = KMC$cluster

KMC$centers[2]
```

```r
# Plot the image with the clusters

dim(healthyClusters) = c(nrow(healthyMatrix), ncol(healthyMatrix))

image(healthyClusters, axes = FALSE, col=rainbow(k))
```

# Apply to a test image

```r
tumor = read.csv("data/tumor.csv", header=FALSE)
```

```r
tumorMatrix = as.matrix(tumor)
tumorVector = as.vector(tumorMatrix)
```

### Apply clusters from before to new image, using the flexclust package

```r
#install.packages("flexclust")
library(flexclust)
```

```r
KMC.kcca = as.kcca(KMC, healthyVector)

# Prediction
tumorClusters = predict(KMC.kcca, newdata = tumorVector)
```

```r
# Visualize the clusters
dim(tumorClusters) = c(nrow(tumorMatrix), ncol(tumorMatrix))

image(tumorClusters, axes = FALSE, col=rainbow(k))
```

We can detect the tumor. It is this abnormal substance here that is highlighted in green that was not present in the healthy MRI image. So we were successfully able to identify, more or less, the geometry of the malignant structure.
