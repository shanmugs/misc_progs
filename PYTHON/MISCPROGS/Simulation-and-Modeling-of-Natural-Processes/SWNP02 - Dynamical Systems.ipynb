```python
# Importing libraries
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Defining parameters
k_a = 1                         # reproduction rate of the antelopes
k_c = 1                         # death rate of cheetahs
k_ac = 0.5                      # death rate of antelopes when the meet cheetahs
k_ca = 0.5                      # reproduction rate of the cheetahs when they meet antelopes

del_t = 0.001                   # Interval taken
t_min = 0                       # Starting time
t_max = 30000                   # Ending time of simulation

t = np.zeros(t_max - t_min)     # Time Vector construction
for m in range(0, len(t)):
    t[m] = m*del_t

P_a = np.zeros(len(t))          # Population of antelope at time t
P_c = np.zeros(len(t))          # Population of cheetahs at time t
dP_a = np.zeros(len(t))
dP_c = np.zeros(len(t))

P_a[0] = 3                      # Population of antelope at time 0
P_c[0] = 5                      # Population of cheetahs at time 0


# Euler's Explicit scheme
for i in range(0, len(t) - 1):
    
    dP_a[i] = k_a*P_a[i] - k_ca*P_c[i]*P_a[i]         # Rate of change of population for antelope
    
    dP_c[i] = k_ac*P_a[i] * P_c[i] - k_c*P_c[i]       # Rate of change of population for cheetah
    
    P_a[i+1] = P_a[i] + del_t*dP_a[i]                 # Population of antelope at time t+1 based on time t
    
    P_c[i+1] = P_c[i] + del_t*dP_c[i]                 # Population of cheetah at time t+1 based on time t

    
# Plotting the population of cheetahs and antelopes
plt.figure(figsize=(20,8))
plt.plot(t, P_a, linewidth=2.0,label="Antelop population")
plt.plot(t, P_c, linewidth=2.0,label="Cheetah population")
plt.xlabel('Time')               # add x-label
plt.ylabel('Population')         # add y-label
plt.title('Prey-Predator Model') # add title
plt.legend(loc='upper right')    # add legend
plt.show()
```
