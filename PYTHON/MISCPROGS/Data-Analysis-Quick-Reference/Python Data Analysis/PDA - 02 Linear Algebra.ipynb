### Fitting to Polynomials with NumPy

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate two random vectors
v1 = np.random.rand(4)
v2 = np.random.rand(4)

# Creates a sequence of equally separated values
sequence = np.linspace(v1.min(), v1.max(), num=len(v1)*10)

# Fit the data to a polynomial of degree 3
coefs = np.polyfit(v1, v2, 3)

# Evaluate the polynomial on the given sequence
polynomial_sequence = np.polyval(coefs, sequence)

# Plot the polynomial curve
plt.plot(sequence, polynomial_sequence, label='Polynomial Fit', color='blue')

# Plot the original data points (vectors)
plt.scatter(v1, v2, color='red', label='Data Points')

# Title and labels
plt.title('Polynomial Fit and Data Points')
plt.xlabel('v1')
plt.ylabel('v2')

# Add legend
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```

### Decomposing a Matrix Using SVD

$$
A = U . Σ . V^T
$$

```python
from scipy.linalg import svd

# Create a matrix
mat=np.array([[5, 3, 1],[5, 3, 0],[1, 0, 5]])
print("Matrix A:\n",mat)

# Perform matrix decomposition using SVD
U, Sigma, V_transpose = svd(mat)
print("\nLeft Singular Matrix:\n",U)
print("\nDiagonal Matrix:\n", Sigma)
print("\nRight Singular Matrix:\n", V_transpose)
```

### Binomial Distribution

$$
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
$$

Where:

- \( P(X = k) \) is the probability of getting exactly \( k \) successes in \( n \) trials.
- \( \binom{n}{k} \) is the binomial coefficient: \( \frac{n!}{k!(n - k)!} \).
- \( p \) is the probability of success on a single trial.
- \( n \) is the number of trials.
- \( k \) is the number of successes.

```python
# Create an numpy vector of size 5000 with value 0
cash_balance = np.zeros(5000)
cash_balance[0] = 500

# Generate random numbers using Binomial
samples = np.random.binomial(9, 0.5, size=len(cash_balance))

# Update the cash balance
for i in range(1, len(cash_balance)):
    if samples[i] < 5:
        cash_balance[i] = cash_balance[i - 1] - 1
    else:
        cash_balance[i] = cash_balance[i - 1] + 1
        
# Plot the updated cash balance
plt.plot(np.arange(len(cash_balance)), cash_balance)
plt.show()
```

The code starts with an initial cash balance of 500 and generates random outcomes using a binomial distribution. Based on each outcome, the cash balance either increases or decreases by 1. Over time, these changes are accumulated, and the evolution of the cash balance is plotted. This can simulate a simplified version of a random walk or a game of chance with wins and losses.

### Normal Distribution

```python
sample_size = 225000                         # Define the sample size

# Generate random values sample using normal distribution
sample = np.random.normal(size=sample_size)  # Generate a sample of random values from a standard normal distribution

# Create Histogram
n, bins, patch_list = plt.hist(sample,                     # Plot a histogram of the sample
                               int(np.sqrt(sample_size)),  # Number of bins is the square root of the sample size
                               density=True)               # Normalize the histogram to form a probability density

# Set parameters
mu, sigma = 0, 1  # Set the mean (mu) and standard deviation (sigma) for the normal distribution
x = bins          # Use the bin edges as the x-axis values for the normal distribution plot
y = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) )  # Calculate the normal distribution curve

# Plot line plot (or bell curve)
plt.plot(x, y,         # Plot the normal distribution curve
         color='red',  # Set the line color to red
         lw=2)         # Set the line width to 2

plt.show()             # Display the plot
```

### Shapiro-Wilk test
The `shapiro()` function will return tuples of two values: teststatistics and p-value. Let's see the following example:

```python
from scipy.stats import shapiro

# Apply Shapiro-Wilk Test
print("Shapiro-Wilk Test for Small Sample: ",shapiro(small_sample))
print("\nShapiro-Wilk Test for Medium Sample:",shapiro(medium_sample))
print("\nShapiro-Wilk Test for Large Sample: ",shapiro(large_sample))
```

In the preceding code block, you can see that the small and large datasets have **p-values greater than 0.05**, so as the null hypothesis has failed to reject it, this means that **the sample looks like a Gaussian or normal distribution**; while for the medium dataset, the p-value is less than 0.05, so the null hypothesis has rejected it, which means the sample does not look like a Gaussian or normal distribution.

```python
large_sample_result = shapiro(large_sample)         # Perform Shapiro-Wilk test on large sample

large_sample_pvalue = large_sample_result.pvalue    # Extract p-value

if large_sample_pvalue > 0.05:                      # Select answer based on the pvalue criteria
    print('The sample looks like a Gaussian or normal distribution.')
else:
    print('The sample does not look like a Gaussian or normal distribution.')
```

### Creating a Masked Array Using the Numpy.ma Subpackage

```python
from scipy.datasets import face
import numpy as np
import matplotlib.pyplot as plt

# Load the face image using the new method
face_image = face()

# Create a random mask array
mask_random_array = np.random.randint(0, 3, size=face_image.shape)

# Create subplots
fig, ax = plt.subplots(nrows=2, ncols=2)

# Display the Original Image
plt.subplot(2,2,1)
plt.imshow(face_image)
plt.title("Original Image")
plt.axis('off')

# Display masked array
masked_array = np.ma.array(face_image, mask=mask_random_array)
plt.subplot(2,2,2)
plt.title("Masked Array")
plt.imshow(masked_array)
plt.axis('off')

# Log operation on original image
plt.subplot(2,2,3)
plt.title("Log Operation on Original")
plt.imshow(np.ma.log(face_image).astype('uint8'))
plt.axis('off')

# Log operation on masked array
plt.subplot(2,2,4)
plt.title("Log Operation on Masked")
plt.imshow(np.ma.log(masked_array).astype('uint8'))
plt.axis('off')

# Display the subplots
plt.show()
```

In the preceding code block, we first loaded the face image from the scipy.misc
subpackage and created a random mask using the `randint()` function. Then, we applied
the random mask on the face image. After this, we applied the log operation on the original
face image and masked face image. Finally, we displayed all the images in 2*2 subplots.
You can also try a range of mask operations on the image from the numpy.ma subpackage.
Here, we are only focusing on the log operation of the masked array.
