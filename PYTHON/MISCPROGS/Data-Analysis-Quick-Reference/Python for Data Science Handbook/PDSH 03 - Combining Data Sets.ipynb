```python
import pandas as pd
import numpy as np
```

```python
class display(object):
    
    """Display HTML representation of multiple objects"""
    
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    
    def __init__(self, *args):
        self.args = args
        
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)
    
    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
```

```python
def make_df(cols, ind):
    
    """Quickly make a DataFrame"""
    
    data = {c: [str(c) + str(i) for i in ind]
            for c in cols}
    
    return pd.DataFrame(data, ind)

# example DataFrame
make_df('ABC', range(3))
```

### Concatenation of NumPy Arrays

```python
x = [1, 2, 3]
y = [4, 5, 6]
z = [7, 8, 9]

np.concatenate([x, y, z])
```

```python
x = [[1, 2],
     [3, 4]]

np.concatenate([x, x], axis=1)
```

### Simple Concatenation with `pd.concat`

```python
ser1 = pd.Series(['A', 'B', 'C'], index=[1, 2, 3])
ser2 = pd.Series(['D', 'E', 'F'], index=[4, 5, 6])

pd.concat([ser1, ser2])
```

```python
df1 = make_df('AB', [1, 2])
df2 = make_df('AB', [3, 4])

display('df1', 'df2', 'pd.concat([df1, df2])')
```

```python
df3 = make_df('AB', [0, 1])
df4 = make_df('CD', [0, 1])

display('df3', 'df4', "pd.concat([df3, df4], axis=1)")
```

```python
x = make_df('AB', [0, 1])
y = make_df('AB', [2, 3])

y.index = x.index  # make duplicate indices!

display('x', 'y', 'pd.concat([x, y])')
```

#### Catching the repeats as an error
If you'd like to simply verify that the indices in the result of pd.concat() do not overlap, you can specify the `verify_integrity` flag. With this set to True, the concatenation will raise an exception if there are duplicate indices. Here is an example, where for clarity we'll catch and print the error message:

```python
try:
    pd.concat([x, y], verify_integrity=True)
    
except ValueError as e:
    print("ValueError:", e)
```

## Categories of Joins

#### One-to-one joins

```python
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})

df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
                    'hire_date': [2004, 2008, 2012, 2014]})

display('df1', 'df2')
```

```python
df3 = pd.merge(df1, df2)
df3
```

#### Many-to-one joins

```python
df4 = pd.DataFrame({'group': ['Accounting', 'Engineering', 'HR'],
                    'supervisor': ['Carly', 'Guido', 'Steve']})

display('df3', 'df4', 'pd.merge(df3, df4)')
```

#### Many-to-many joins

```python
df5 = pd.DataFrame({'group': ['Accounting', 'Accounting',
                              'Engineering', 'Engineering', 'HR', 'HR'],
                    'skills': ['math', 'spreadsheets', 'coding', 'linux',
                               'spreadsheets', 'organization']})

display('df1', 'df5', "pd.merge(df1, df5)")
```

### Specification of the Merge Key

```python
display('df1', 'df2', "pd.merge(df1, df2, on='employee')")
```

#### The left_on and right_on keywords

```python
df3 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'salary': [70000, 80000, 120000, 90000]})

display('df1', 'df3', 'pd.merge(df1, df3, left_on="employee", right_on="name")')
```

```python
pd.merge(df1, df3, left_on="employee", right_on="name").drop('name', axis=1)
```

#### The left_index and right_index keywords

```python
display('df1', 'df2')
```

```python
display('df1a', 'df2a', "pd.merge(df1a, df2a, left_index=True, right_index=True)")
```

```python
display('df1a', 'df2a', 'df1a.join(df2a)')
```

```python
display('df1a', 'df3', "pd.merge(df1a, df3, left_index=True, right_on='name')")
```

### Specifying Set Arithmetic for Joins

```python
df6 = pd.DataFrame({'name': ['Peter', 'Paul', 'Mary'],
                    'food': ['fish', 'beans', 'bread']},
                   columns=['name', 'food'])

df7 = pd.DataFrame({'name': ['Mary', 'Joseph'],
                    'drink': ['wine', 'beer']},
                   columns=['name', 'drink'])

display('df6', 'df7', 'pd.merge(df6, df7)')
```

```python
pd.merge(df6, df7, how='inner')
```

```python
display('df6', 'df7', "pd.merge(df6, df7, how='outer')")
```

```python
display('df6', 'df7', "pd.merge(df6, df7, how='left')")
```

### Overlapping Column Names: The suffixes Keyword

```python
df8 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'rank': [1, 2, 3, 4]})

df9 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'rank': [3, 1, 4, 2]})

display('df8', 'df9', 'pd.merge(df8, df9, on="name")')
```

```python
display('df8', 'df9', 'pd.merge(df8, df9, on="name", suffixes=["_L", "_R"])')
```

```python
pop = pd.read_csv('data/state-population.csv')
areas = pd.read_csv('data/state-areas.csv')
abbrevs = pd.read_csv('data/state-abbrevs.csv')

display('pop.head()', 'areas.head()', 'abbrevs.head()')
```

```python
merged = pd.merge(pop, abbrevs, how='outer', left_on='state/region', right_on='abbreviation')

merged.head()
```

```python
merged = merged.drop(columns=['abbreviation']) # drop duplicate info

merged.head()
```

```python
# Let's look for rows with nulls:

merged.isnull().any()
```

```python
merged.isna().sum()
```

Some of the population info is null; let's figure out which these are!

```python
merged[merged['population'].isnull()].head()
```

```python
merged.loc[merged['state'].isnull(), 'state/region'].unique()
```

```python
merged.loc[merged['state/region'] == 'PR', 'state'] = 'Puerto Rico'
merged.loc[merged['state/region'] == 'USA', 'state'] = 'United States'

merged.isnull().any()
```

```python
# Let's confirm the change:
row = merged.iloc[1874]
print(row)
```

```python
final = pd.merge(merged, areas, on='state', how='left')
final.head()
```

```python
final.isnull().any()
```

```python
final['state'][final['area (sq. mi)'].isnull()].unique()
```

```python
final.dropna(inplace=True)
final.head()
```

```python
data2010 = final.query("year == 2010 & ages == 'total'")
data2010.head()
```

```python
data2010.set_index('state', inplace=True)
density = data2010['population'] / data2010['area (sq. mi)']

density.sort_values(ascending=False, inplace=True)
density.head()
```
