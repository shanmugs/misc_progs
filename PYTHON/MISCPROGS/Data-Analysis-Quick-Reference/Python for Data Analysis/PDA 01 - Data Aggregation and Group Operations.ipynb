# Random

```python
frame = pd.DataFrame({"data1": np.random.standard_normal(1000),
                      "data2": np.random.standard_normal(1000)})
frame.head()
```

```python
quartiles = pd.cut(frame["data1"], 4)

quartiles.head(10)
```

```python
def get_stats(group):
    return pd.DataFrame(
        {"min": group.min(), "max": group.max(),
        "count": group.count(), "mean": group.mean()}
    )

grouped3 = frame.groupby(quartiles, observed=False)   ############# <=========== NOVO PANDAS EXIGE: ', observed=False'

grouped3.apply(get_stats)
```

```python
grouped3.agg(["min", "max", "count", "mean"])
```

```python
quartiles_samp = pd.qcut(frame["data1"], 4, labels=False)
quartiles_samp.head()
```

```python
def get_stats(group):
    return pd.DataFrame(
        {"min": group.min(), "max": group.max(),
        "count": group.count(), "mean": group.mean()}
    )
    
grouped4 = frame.groupby(quartiles_samp)
grouped4.apply(get_stats)
```

# Cards

```python
suits = ["H", "S", "C", "D"]  # Hearts, Spades, Clubs, Diamonds
card_val = (list(range(1, 11)) + [10] * 3) * 4
base_names = ["A"] + list(range(2, 11)) + ["J", "K", "Q"]
cards = []

for suit in suits:
    cards.extend(str(num) + suit for num in base_names)

deck = pd.Series(card_val, index=cards)

deck.head(13)
```

```python
def draw(deck, n=5):         # Get the 'deck' sequence, 5 cards
    return deck.sample(n)    # Randomly
    
draw(deck)
```

```python
# Group by Suits

def get_suit(card):
    # last letter is suit
    return card[-1]

deck.groupby(get_suit).apply(draw, n=2)
```

```python
# get_suit function extracts the suit from each card string ("AH" -> "H") 
# group_keys=False prevents the grouping keys from becoming part of the index.

deck.groupby(get_suit, group_keys=False).apply(draw, n=2)
```

# Stock

```python
close_px = pd.read_csv("stock_px.csv", parse_dates=True,
                       index_col=0)
close_px.info()
```

```python
close_px.tail(4)
```

```python
# Computes the correlation (corrwith) of each stock's returns (group) with the returns of the S&P 500 index (group["SPX"])
def spx_corr(group):
    return group.corrwith(group["SPX"])

# pct_change() computes the percentage change in prices, and dropna() removes NaN
rets = close_px.pct_change().dropna()

# Get Year in the Timeseries
def get_year(x):
    return x.year

by_year = rets.groupby(get_year)
by_year.apply(spx_corr)
```

```python
# Calculates the correlation between the two Series (columns) of stock prices
def corr_aapl_msft(group):
    return group["AAPL"].corr(group["MSFT"])
    
by_year.apply(corr_aapl_msft)
```

```python
import statsmodels.api as sm

def regress(data, yvar=None, xvars=None):
    Y = data[yvar]
    X = data[xvars]
    X["intercept"] = 1.
    result = sm.OLS(Y, X).fit()
    return result.params
```

```python
by_year.apply(regress, yvar="AAPL", xvars=["SPX"])
```

```python
def times_two(group):
    return group * 2
    
g.transform(times_two)
```

```python
def get_ranks(group):
    return group.rank(ascending=False)
    
g.transform(get_ranks)
```
