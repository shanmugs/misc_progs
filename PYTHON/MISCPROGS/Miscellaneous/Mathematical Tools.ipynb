# Derivative

Centered Derivative:

$$
f'(x) = \frac{f(x+dx) - f(x-dx)}{2dx}
$$

```python
# f(x) = x^2

def f(x):                # Define Function
    return (x)**2

def derivative(x, dx):   # Define Derivative
    d = (f(x+dx) - f(x-dx)) / (2*dx) 
    return d

x = 3
dx = 1e-5                # 0.00001

der = derivative(x, dx)
ana = 2 * x

print("Numerical:", der)
print("Analytical:", ana)
```

# Fitting Polynomial Regression

```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Plot 3 points

```python
import matplotlib.pyplot as plt

# Create X and Y lists
X = [0, 1, 4]
y = [0, 3, 2]

# Plot scatter diagram
plt.scatter(X, y, color='red')
plt.xlim(-2, 8)                 # Set X-axis limits
plt.ylim(-2, 8)                 # Set Y-axis limits

# Add lines for x=0 and y=0
plt.axhline(0, color='black', linestyle='--')  # Horizontal line for y=0
plt.axvline(0, color='black', linestyle='--')  # Vertical line for x=0

plt.title('Polynomial Regression')
plt.xlabel('X-Axis')
plt.ylabel('y-Axis')

plt.show()
```

### Find the Polinomial with degree 3 that fit with these 3 points

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Prepare dataset
data = pd.DataFrame({"X": [0, 1, 4], "y": [0, 3, 2]})
X = data[['X']] 
y = data[['y']]

# Apply Polynomial Features
polynomial_reg = PolynomialFeatures(degree=3)
X_polynomial = polynomial_reg.fit_transform(X)

# Apply Linear Regression Model
linear_reg = LinearRegression()
linear_reg.fit(X_polynomial, y)

# Generate a range of X values for smooth plotting
X_range = pd.DataFrame(np.linspace(X.min(), X.max(), 100), columns=X.columns)  # Ensure feature names match
X_range_polynomial = polynomial_reg.transform(X_range)

# Make predictions for the smooth X range
predictions = linear_reg.predict(X_range_polynomial)

# Plot the results
plt.scatter(X, y, color='red')
plt.plot(X_range, predictions, color='blue')
plt.xlim(-2, 8)  # Set X-axis limits
plt.ylim(-2, 8)  # Set Y-axis limits

# Add lines for x=0 and y=0
plt.axhline(0, color='black', linestyle='--')  # Horizontal line for y=0
plt.axvline(0, color='black', linestyle='--')  # Vertical line for x=0

plt.title('Polynomial Regression')
plt.xlabel('X-Axis')
plt.ylabel('y-Axis')

plt.show();
```

# Integral

### Full Simpson's Rule Formula

The approximation over the entire interval \([a, b]\) is obtained by summing up the contributions of all pairs of subintervals:

$$
\int_a^b f(x) \, dx \approx \frac{h}{3} \left[ f(x_0) + 4 \sum_{\text{odd } i} f(x_i) + 2 \sum_{\text{even } i} f(x_i) + f(x_n) \right]
$$

where:
- \( f(x_0) \) and \( f(x_n) \) are the values at the endpoints of the interval.
- The odd-indexed terms (\( x_1, x_3, \ldots \)) are weighted by 4.
- The even-indexed terms (\( x_2, x_4, \ldots \)) are weighted by 2.

This approach is much more accurate than using straight-line approximations (like in the trapezoid rule), as it accounts for the curvature of the function within each pair of subintervals.


```python
# Define the function
def f(x):
    return -0.5 * x**3 + 1.5 * x**2 + 2 * x  # The function to be integrated

# Simpson's Rule implementation
def simpsons_rule(f, a, b, n):
    
    if n % 2 != 0:   # Ensure n is even
        raise ValueError("Number of intervals n must be even for Simpson's rule.")
    
    h = (b - a) / n  # Calculate width of each subinterval
    x = a            # Starting point

    # Initialize sums for Simpson's Rule
    integral = f(a) + f(b)  # f(x0) and f(xn)
    odd_sum = 0             # Sum for odd indices
    even_sum = 0            # Sum for even indices

    # Iterate through subintervals
    for i in range(1, n):
        x += h
        if i % 2 == 0:  # Even index
            even_sum += f(x)
        else:  # Odd index
            odd_sum += f(x)

    # Calculate final integral value
    integral += 4 * odd_sum + 2 * even_sum
    integral *= h / 3
    
    return integral

# Define integration limits and number of subintervals
a = 1   # Lower limit
b = 4   # Upper limit
n = 10  # Number of intervals (even)

# Calculate the integral using Simpson's Rule
integral = simpsons_rule(f, a, b, n)

# Print the result
print(f"The integral of f(x) from x = {a} to x = {b} using Simpson's Rule is: {integral:.4f}")
```

# Ordinary Diferential Equation

To solve the equation:
$$
545 = 49t + 245 e^{-\frac{t}{5}}
$$

```python
import numpy as np
from scipy.optimize import fsolve

# Define the equation as a function
def equation(t):
    return 49 * t + 245 * np.exp(-t / 5) - 545

# Use fsolve to find the root (approximate solution for t)
t_solution = fsolve(equation, 0)  # 0 is the initial guess

print(f"The solution for t is approximately: {t_solution[0]}")
```
